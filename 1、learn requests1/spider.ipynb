## 3、案例二：百度翻译

```python
from requests import *

# 输入dog
url = "https://fanyi.baidu.com/sug"

s = input("请输入你要翻译的英文单词")

data = {
    "kw" : s
}

# 发送post请求，数据必须放在字典中，通过data参数进行传递。
resp = post(url, data = data)

print(resp.text)

# 将服务器里的内容直接转变成JSON ==> 也就是字典dict
print(resp.json())

# 关掉resp
resp.close()
```

## 案例三：豆瓣排行榜

```python
import requests

# https://movie.douban.com/typerank?type_name=%E5%96%9C%E5%89%A7&type=24&interval_id=100:90
# 应该是：https://movie.douban.com/j/chart/top_list?type=24&interval_id=100%3A90&action=&start=0&limit=20
# 原来很长，我们用字典封装方式
url = "https://movie.douban.com/j/chart/top_list"

params = {
    # "type_name" : "戏剧",
    "type": "24",
    "interval_id": "100:90",
    "action": "",
    "start": 0,
    "limit": 20
}

dic = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36"
}


resp = requests.get(url = url, params = params, headers = dic)
print(resp.request.url)
print(resp.text)

print(resp.json())

# 关掉resp
resp.close()
```

## 案例四：北京新发地菜价



这个需要我们点击查询之后才能使用。
这是一个异步渲染的题目，网页源代码中是没有的。

```python
import requests
import csv

# url = "http://www.xinfadi.com.cn/priceDetail.html"
# 地址不要搞错，是解包工具中的Fetch/XHR中的http://www.xinfadi.com.cn/getPriceData.html
url = "http://www.xinfadi.com.cn/getPriceData.html"

# 网站的payload默认是空。
# payload如果是from data，直接挂字典就可以了。
num = input("请问你查询的数量是多少：")

data = {
    "limit": num,
    "current": "",
    "pubDateStartTime": "",
    "pubDateEndTime": "",
    "prodPcatid": "",
    "prodCatid": "",
    "prodName": "",
}

header = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36"
}

# 发送post请求，数据必须放在字典中，通过data参数进行传递。
resp = requests.post(url, data=data, headers=header)

# 用来获取json数据的，前提是：必须服务器返回的是json格式，如果不是，不能这么用。
dic = resp.json()

f = open("data.csv", mode="w",encoding="utf-8",newline="")
csvwriter = csv.writer(f)
csvwriter.writerow(["prodName", "prodCat", "lowPrice", "avgPrice", "highPrice", "place", "pubDate"])

# 获取JSON数据的商品列表
product_list = dic.get('list', [])

# 遍历商品列表
for product in product_list:
    # 提取所需字段
    prodName = product.get('prodName', '')
    prodCat = product.get('prodCat', '')
    lowPrice = product.get('lowPrice', '')
    avgPrice = product.get('avgPrice', '')
    highPrice = product.get('highPrice', '')
    place = product.get('place', '')
    pubDate = product.get('pubDate', '')

    # 将提取到的字段写入CSV文件
    csvwriter.writerow([prodName, prodCat, lowPrice, avgPrice, highPrice, place, pubDate])

# 关闭文件
f.close()
print("over!")

resp.close()
```
